```python

df['InvoiceNo'].str.contains('C').count()
```

```python
dfn = df[~df['InvoiceNo'].str.contains('C', na=False)] # اضفت علامه ~ ف جاب العكس الي هيه المنتجات الي تم بيعها بشكل كامل
```

```python
dfn
```

```python
dfn['TotalPrice'] = df['Quantity'] * df['UnitPrice']
```

```python
dfn
```

```python
dfn['InvoiceDate'].max()
```

```python
today_date = dt.datetime(2011,12,11)
```

```python
rfm = dfn.groupby('CustomerID').agg({'InvoiceDate': lambda date: (today_date - date.max()).days,
                                    'InvoiceNo' : lambda num: num.nunique(),
                                    'TotalPrice': lambda TotalPrice: TotalPrice.sum()})
rfm
```

```python
rfm.columns = ['Recency','Frequency','Monitry']
rfm
```

```python
rfm = rfm[rfm['Monitry'] > 0 ]
rfm
```

```python
rfm.describe().T
```

```python
rfm['Recency_score'] = pd.qcut(rfm['Recency'] , 5, labels= [5,4,3,2,1]) 
```

```python
rfm['Frequency_score'] = pd.qcut(rfm['Frequency'].rank(method='first'),5, labels=[1,2,3,4,5]) 
```

```python
rfm['Monitry_score'] = pd.qcut(rfm['Monitry'] , 5, labels= [1,2,3,4,5]) 
```

```python
rfm.sample(10)
```

```python
rfm['RfmScore'] = (rfm['Recency_score'].astype('str') + rfm['Frequency_score'].astype('str'))
```

```python
rfm
```

```python
# Segmenting customer using RFM Ccore

Seg_map = {

r'[1-2][1-2]' : 'hibernating',
r'[1-2][2-3]' : 'at_Risk',
r'[1-2]5' : 'cant_loose',
r'3[1-2]' : 'about_to_sleep',
r'33' : 'need_attention',
r'[3-4][4-5]' : 'loyal_customer',
r'41' : 'promising',
r'51' : 'new_customer',
r'[4-5][2-3]' : 'pottntial_loyalists',
r'5[4-5]' : 'champions'

}

rfm['segment'] = rfm['RfmScore'].replace(Seg_map, regex=True)
```

```python
rfm.sample(10)
```

```python
rfm[['segment','Recency','Frequency','Monitry']].groupby('segment').agg(['mean','count','max']).round(1)
```
